## multiline %%sql to pandas.DataFrame()

```python
%%sql df2 <<
select * 
    from movie_gross 
    where genre = 'Adventure'
    limit 5;
```

```python
df2
```

```python
type(df2)
```

```python
df2.info()
```

```python
df2['release_date']=df2['release_date'].astype('datetime64')
```

```python
df2.info()
```

```python
df2[df2['release_date'].dt.year < 1960]
```
